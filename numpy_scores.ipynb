```python
import numpy as np

# Task 1 — Generate and Inspect Data
np.random.seed(42)

# generate 5x4 array with values between 50 and 100
scores = np.random.randint(50, 101, size=(5,4))

print("Scores Matrix:")
print(scores)

# score of 3rd student in 2nd subject
print("\nScore of 3rd student in 2nd subject:")
print(scores[2,1])

# scores of last 2 students
print("\nScores of last 2 students:")
print(scores[-2:])

# first 3 students, subjects 2 and 3
print("\nScores of first 3 students in subjects 2 and 3:")
print(scores[:3,1:3])


# Task 2 — Broadcasting Operations

# column-wise mean
column_mean = np.round(scores.mean(axis=0),2)
print("\nColumn-wise mean:")
print(column_mean)

# add curve using broadcasting
curve = np.array([5,3,7,2])
curved_scores = scores + curve

# ensure max score = 100
curved_scores = np.clip(curved_scores, None, 100)

print("\nCurved Scores:")
print(curved_scores)

# row-wise max
row_max = curved_scores.max(axis=1)
print("\nRow-wise max:")
print(row_max)


# Task 3 — Normalize and Identify

# row-wise min and max
row_min = curved_scores.min(axis=1, keepdims=True)
row_max = curved_scores.max(axis=1, keepdims=True)

# min-max normalization
normalized = (curved_scores - row_min) / (row_max - row_min)

print("\nNormalized scores:")
print(normalized)

# index of highest value in normalized array
max_index = np.unravel_index(np.argmax(normalized), normalized.shape)

print("\nIndex of highest normalized value (student,row , subject,column):")
print(max_index)

# scores strictly above 90
above_90 = curved_scores[curved_scores > 90]

print("\nScores above 90:")
print(above_90)
```
